# Wind Plots Example (Barbs and Quivers)\n\nThis notebook demonstrates how to use `monet-plots` to create wind barb and wind quiver plots using Cartopy.

```python
import numpy as np\nimport xarray as xr\nimport cartopy.crs as ccrs\nimport cartopy.feature as cfeature\nfrom monet_plots.plots.wind_barbs import WindBarbsPlot\nfrom monet_plots.plots.wind_quiver import WindQuiverPlot\nimport matplotlib.pyplot as plt
```

```python
# Generate sample wind data (u, v components)\nnp.random.seed(42)\nlon = np.arange(-10, 2)\nlat = np.arange(20, 40, 2)\nU = np.random.rand(len(lat), len(lon)) * 10 - 5  # u component\nV = np.random.rand(len(lat), len(lon)) * 10 - 5  # v component\n\n# Create an xarray DataArray for coordinates (optional, but good practice for spatial data)\nds = xr.Dataset(\n    {\n        'u': (('lat', 'lon'), U),\n        'v': (('lat', 'lon'), V),\n    },\n    coords={'lon': lon, 'lat': lat}\n)
```

```python
# --- Wind Barbs Plot ---\n# Create a WindBarbsPlot instance\nbarbs_plot = WindBarbsPlot(figsize=(10, 8), projection=ccrs.PlateCarree())\n\n# Plot wind barbs\nbarbs_plot.plot(x=ds['lon'], y=ds['lat'], u=ds['u'], v=ds['v'], transform=ccrs.PlateCarree())\n\n# Add geographical features\nbarbs_plot.ax.add_feature(cfeature.COASTLINE)\nbarbs_plot.ax.add_feature(cfeature.BORDERS, linestyle=':')\nbarbs_plot.ax.add_feature(cfeature.LAND, facecolor='lightgray')\nbarbs_plot.ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)\n\n# Add title\nbarbs_plot.ax.set_title('Wind Barbs Plot Example')\n\n# Save the plot\nbarbs_plot.save('wind_barbs_example.png')
```

```python
# --- Wind Quiver Plot ---\n# Create a WindQuiverPlot instance\nquiver_plot = WindQuiverPlot(figsize=(10, 8), projection=ccrs.PlateCarree())\n\n# Plot wind quivers\nquiver_plot.plot(x=ds['lon'], y=ds['lat'], u=ds['u'], v=ds['v'], transform=ccrs.PlateCarree())\n\n# Add geographical features\nquiver_plot.ax.add_feature(cfeature.COASTLINE)\nquiver_plot.ax.add_feature(cfeature.BORDERS, linestyle=':')\nquiver_plot.ax.add_feature(cfeature.LAND, facecolor='lightgray')\nquiver_plot.ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)\n\n# Add title\nquiver_plot.ax.set_title('Wind Quiver Plot Example')\n\n# Save the plot\nquiver_plot.save('wind_quiver_example.png')
```
