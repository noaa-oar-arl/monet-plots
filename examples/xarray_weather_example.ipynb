# Xarray Weather Data Example

This notebook demonstrates how to use `monet-plots` to create a spatial plot from xarray's toy weather data.

```python
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from monet_plots.plots.spatial import SpatialPlot
```

```python
# Load the xarray toy weather dataset
air = xr.tutorial.load_dataset("air_temperature")
air
```

```python
# Create a spatial plot instance
plot = SpatialPlot(projection=ccrs.PlateCarree(), figsize=(12, 8))

# Select a time slice and plot the data
air_slice = air.air.isel(time=0)
plot.plot(air_slice)

# Add a title
plot.ax.set_title("Surface Air Temperature")

# Save the plot
plot.save("xarray_weather_example.png")
```
